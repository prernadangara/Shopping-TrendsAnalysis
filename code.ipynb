```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/content/shopping_trends_updated.csv")

# Data Preprocessing
df.dropna(inplace=True)  # Removing missing values

# Visualization: Purchase Amount Distribution
plt.figure(figsize=(8,5))
sns.histplot(df["Purchase Amount (USD)"], bins=30, kde=True, color="blue")
plt.xlabel("Purchase Amount (USD)")
plt.ylabel("Frequency")
plt.title("Distribution of Purchase Amounts")
plt.show()

# Most Purchased Categories
plt.figure(figsize=(8,5))
sns.countplot(y=df["Category"], order=df["Category"].value_counts().index, palette="coolwarm")
plt.xlabel("Count")
plt.ylabel("Category")
plt.title("Most Purchased Categories")
plt.show()

# Seasonal Shopping Trends
plt.figure(figsize=(8,5))
sns.countplot(x=df["Season"], order=df["Season"].value_counts().index, palette="magma")
plt.xlabel("Season")
plt.ylabel("Number of Purchases")
plt.title("Purchases by Season")
plt.show()
```
